# Logistic Map#

```python
# fn을 구하는 함수를 작성하여라.
def logistic (r,s) :
 fn=[]
 x = 0
 while(x<100):
  s = r*s*(1-s)
  x = x+1
  fn.append(s)
 return fn
```

```python
# 작성한 함수를 이용하여 다음의 조건에서 n=1에서 100까지 로지스틱 맵의 값을 구하고 
# 그래프로 그려라. 즉, x축은  n, y축은 fn인 그래프를 그려라.

def logistic (r,s) :
 fn=[]
 x = 0
 while(x<100):
  s = r*s*(1-s)
  x = x+1
  fn.append(s)
 return fn
plt.figure(figsize=(16,8))
plt.subplot(231)
plt.plot(logistic(0.5,0.5))
plt.subplot(232)
plt.plot(logistic(0.95,0.5))
plt.subplot(233)
plt.plot(logistic(2,0.5))
plt.subplot(234)
plt.plot(logistic(3.2,0.5))
plt.subplot(235)
plt.plot(logistic(3.8,0.5))
plt.subplot(236)
plt.plot(logistic(3.8,0.501))
```

# sinc function

```python
#싱크 함수를 그려보아라. 
a=np.arange(-50,50,0.01)
b=[np.sin(x)/x for x in a]
plt.figure(figsize=(8,8))
plt.plot(a,b)
```

```python
# 싱크함수와 싱크함수를 테일러 전개한 식을 그래프를 통해 비교하여라.
def facto(x) :
  f,n = 1,1
  while(n<=x) :
    f=f*n
    n=n+1
  return f

def f(s) : 
  A=[]
  n=1
  x=0
  while(n<=s) :
    x=x+(s**(2*n)/((facto(2*n+1)*(-1)**n)))
    n=n+1
    A.append(x)
  return A
a=[q for q in range(0,100)]
b=[np.sin(p)/p for p in a]
plt.figure(figsize=(16,8))
plt.plot(a,b,label='sinc(x)',color='b')
plt.plot(f(50),label='Taylor', color='r')
```
